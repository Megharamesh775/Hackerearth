Classes are imbalanced , we need to oversample the minority classes and undersample the majority classes

### SAMPLING

```python
from sklearn.utils import resample
aai_df=train_df[train_df['LICENSE STATUS']== 0]
aac_df=train_df[train_df['LICENSE STATUS']== 1]  
rev_df=train_df[train_df['LICENSE STATUS']== 2]    
rea_df=train_df[train_df['LICENSE STATUS']== 3]
inq_df=train_df[train_df['LICENSE STATUS']== 4]

aai_downsample=resample(aai_df,replace=False,n_samples=350)
aac_downsample=resample(aac_df,replace=False,n_samples=300)
rea_downsample=resample(rea_df,replace=True,n_samples=350)
inq_downsample=resample(inq_df,replace=True,n_samples=300)
```

```python
new_df=pd.concat([aai_downsample,aac_downsample,rev_df,rea_downsample,inq_downsample])
from sklearn.utils import shuffle
snew_df=shuffle(new_df)
```

```python
snew_df.head()
```

```python
plt.title('Distribution of target Variable')
sns.countplot(snew_df['LICENSE STATUS'])
plt.show()
```

```python
#label encoder
from sklearn.preprocessing import LabelEncoder
def le(df):
    for i in df.columns:
        if df[i].dtypes=='object':
            print(i,'encoded')
            mask= ~df[i].isnull()
            label_enc=LabelEncoder()
            try:
                df[i][mask]=label_enc.fit_transform(df[i][mask])
                df[i]=df[i].astype(int)
            except Exception as e:
                print(e)
    return df
```

```python
snew_df=le(snew_df)
```

```python
X=snew_df.drop(['LICENSE STATUS'],axis=1)
y=snew_df['LICENSE STATUS']
```

### Train-test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def cm(actual,predicted):
    con=confusion_matrix(actual,predicted)
    sns.heatmap(con,annot=True,fmt='.2f')
    plt.ylabel('predicted')
    plt.xlabel('Actual')
    plt.show()
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
report=classification_report(y_test,y_pred)
print(report)
acc_n=round(accuracy_score(y_test,y_pred)*100,2)
```

```python
cm(y_test,y_pred)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
clf=DecisionTreeClassifier()
param=[{'max_features':['log2','sqrt','auto'],
       'criterion':['entropy','gini'],
       'max_depth':[2,3,5,10,50],
       'min_samples_split':[2,3,50,100],
       'min_samples_leaf':[1,5,8,10]}]
grid=GridSearchCV(estimator=clf,param_grid=param,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
clf=grid.best_estimator_


y_pred=clf.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_d=round(accuracy_score(y_test,y_pred)*100,2)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
params=[{'n_estimators':[4,6,9,10,15],
         'max_features':['log2','sqrt','auto'],
       'criterion':['entropy','gini'],
       'max_depth':[2,3,5,10],
       'min_samples_split':[2,3,50],
       'min_samples_leaf':[1,5,8]}]
grid=GridSearchCV(estimator=rf,param_grid=params,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
rf=grid.best_estimator_


y_pred=rf.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_r=round(accuracy_score(y_test,y_pred)*100,2)
```

### SVM Classifier

```python
from sklearn import svm
svc=svm.SVC()

params=[{'C':[1,10,100,1000],'kernel':['linear']},
       {'C':[1,10,100,1000],'kernel':['rbf'],'gamma':[0.001,0.0001]}]

grid=GridSearchCV(estimator=svc,param_grid=params,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
svc=grid.best_estimator_


y_pred=svc.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_s=round(accuracy_score(y_test,y_pred)*100,2)
```

### K- Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()

params=[{'n_neighbors':[3,7,5,10],'weights':['uniform','distance']
        ,'algorithm':['auto','ball_tree','kd_tree','brute'],
        'leaf_size':[10,20,30,50]}]
grid=GridSearchCV(estimator=knn,param_grid=params,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
knn=grid.best_estimator_


y_pred=knn.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_k=round(accuracy_score(y_test,y_pred)*100,2)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)

y_pred=log.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_l=round(accuracy_score(y_test,y_pred)*100,2)
```

### XGBOOST

```python
from xgboost import XGBClassifier
xgb=XGBClassifier(learning_rate=0.02,n_estimators=1000,objective='multi:softmax',silent=True,nthread=1,num_class=5)

params=[{'min_child_weight':[1,5,10],
        'gamma':[0.1,0.5,1,1.5,2,5],
        'subsample':[0.6,0.8,1.0],
        'max_depth':[3,5,6]}]
grid=GridSearchCV(estimator=xgb,param_grid=params,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
xgb=grid.best_estimator_


y_pred=xgb.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_x=round(accuracy_score(y_test,y_pred)*100,2)
```

### ADABOOST

```python
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()

params=[{'n_estimators':[4,6,9,10,15,30],
        'random_state':np.random.randint(5,30,size=5)}]

grid=GridSearchCV(estimator=ada,param_grid=params,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
ada=grid.best_estimator_


y_pred=ada.predict(X_test)

report=classification_report(y_test,y_pred)
print(report)
cm(y_test,y_pred)
acc_a=round(accuracy_score(y_test,y_pred)*100,2)
```

### Comparison of the models

```python
models=pd.DataFrame({'Model':['Logistic Regression','Naive bayes','Decision Tree','Random Forest',
                             'SVM','KNN','XGboost','Adaboost'],
                    'Score':[acc_l,acc_n,acc_d,acc_r,acc_s,acc_k,acc_x,acc_a]})
models.sort_values(by='Score',ascending=False)
```

We can observe that the best fit model to our dataset is XGboost with 94% accuracy followed by random forest and decision tree classifier.
